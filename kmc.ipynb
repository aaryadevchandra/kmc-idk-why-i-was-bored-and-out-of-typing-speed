```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
def distance(pt1, pt2):
    return np.sqrt((pt1[1]-pt2[1])**2 + (pt1[0] - pt2[0]) ** 2)
```

```python
points, _ = make_blobs(n_samples=100, shuffle=True)
```

```python
plt.title('Displaying Random Points')
plt.plot(points[:, 0], points[:, 1], 'rx')
plt.xlim()
plt.ylim()
plt.grid()
plt.show()
```

```python
z = {'c1': [], 'c2': []}
```

```python
# geenrating random center points

k = 2
centers = []
for _ in range(k):
    centers.append(np.array([np.random.randint(-10, 10), np.random.randint(-10, 10)]))
centers = np.array(centers)
```

```python
centers # print karke dekhing
```

```python
for point in points:
    if distance(point, centers[0]) < distance(point, centers[1]):
        z['c1'].append(point)
    else:
        z['c2'].append(point)
        
```

```python
plt.grid()
plt.plot(np.array(z['c1'])[:, 0], np.array(z['c1'])[:, 1], 'bx')
plt.plot(centers[0][0], centers[0][1], 'bo')

plt.plot(np.array(z['c2'])[:, 0], np.array(z['c2'])[:, 1], 'gx')
plt.plot(centers[1][0], centers[1][1], 'go')
plt.xlim(-15, 15)
plt.ylim(-15, 15)
plt.show()
```

```python
def train_k_means_clustering(z, centers, points, epochs=100):

    for _ in range(epochs):
    
        for point in points:
            if distance(point, centers[0]) < distance(point, centers[1]):
                # print(f'Going to point 1')
                z['c1'].append(point)
            else:
                # print(f'Going to point 2')
                z['c2'].append(point)


        # new centers
        centers[0] = [np.mean(np.array(z['c1'])[:, 0]), np.mean(np.array(z['c1'])[:, 1])]
        centers[1] = [np.mean(np.array(z['c2'])[:, 0]), np.mean(np.array(z['c2'])[:, 1])]

        # displyaing the new points assigned to their new centers
        plt.grid()
        plt.plot(np.array(z['c1'])[:, 0], np.array(z['c1'])[:, 1], 'bx')
        plt.plot(centers[0][0], centers[0][1], 'bo')
        
        plt.plot(np.array(z['c2'])[:, 0], np.array(z['c2'])[:, 1], 'gx')
        plt.plot(centers[1][0], centers[1][1], 'go')
        plt.xlim(-15, 15)
        plt.ylim(-15, 15)
        plt.show()

            
```

```python
train_k_means_clustering(z, centers, points)
```
